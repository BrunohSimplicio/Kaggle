Pega a media de Idade(Age) para os valores nulos 

```python
media_idade = train_titanic['Age'].median()
media_idade
```

Media de Idade nos Valores nulos

```python

train_titanic['Age'] = train_titanic['Age'].fillna(media_idade)
test_titanic ['Age'] = test_titanic ['Age'].fillna(media_idade)
```

```python
train_titanic['Sex'].value_counts()
```

```python
test_titanic['Sex'].value_counts()
```

```python
train_titanic['Sexo_Binario'] = (train_titanic['Sex'] == 'female').astype(int)
train_titanic
```

```python
test_titanic['Sexo_Binario'] = (test_titanic['Sex'] == 'female').astype(int)
test_titanic
```

```python
previsoes = ['Pclass', 'Sexo_Binario', 'Age']
X_train = train_titanic[previsoes].values
X_test  = test_titanic [previsoes].values
y_train = train_titanic['Survived'].values
X_train[:5]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```
